## Section 6. Outlier detection (using Z-score > 2.5)

Outliers get removed only for **numeric features**, such as credit limits, ages, bill amount, and payment amounts.

Categorical features(`SEX`, `EDUCATION`, `MARRIAGE`) were excluded because their values are coded not values.

Uzing the Z-score rule:
$$
z_i = \frac{x_i - \mu_x}{\sigma_x}
$$
any observation with
$$
|z_i| > 2.5
$$  
then (x-i) is considered an outlier and removed.

This method ensures the dataset focused on typical behavior and avoids the influence of the extreme numeric values.

---

```python
numeric_cols_outliers = [
    'LIMIT_BAL','AGE',
    'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
    'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6'
]


initial_count = len(df)

for col in numeric_cols_outliers:
    z = (df[col] - df[col].mean()) / df[col].std()
    df = df[np.abs(z) <= 2.5]  # keep only values within ±2.5 std deviations

final_count = len(df)
removed_count = initial_count - final_count
print(f"Initial rows: {initial_count}, Final rows: {final_count}, Removed rows: {removed_count}")
```

## Section 7. Normalization

After cleaning and removing outliers, all numeric features are **normalized** so that their mean is `0` and stadard deviation is `1` using:
$$
x' = \frac{x - \mu_x}{\sigma_x}
$$

**Why this method**:

- It puts all variables on the same scale, so big fearures like(`LIMIT_BAL`) don't  drown out small ones like (`AGE`).
- Models like Logistic Regression and SVM run more smoothly and reliably.
---

There's another way to normalize data, called min-max normalization:
$$
x' = 2 \left( \frac{x - \min(x)}{\max(x) - \min(x)} - 0.5 \right)
$$
Both methods give dimensionless scalled features.
**Z-score normalization** was chosen here because it preserve Gaussian-like distribution and is widely used in financial numeric data.

---

```python
# Normalization of numeric features (Z-score standardization)
num_features = [
    'LIMIT_BAL','AGE',
    'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
    'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6'
]

# Z-score normalization
for col in num_features:
    df[col] = (df[col] - df[col].mean()) / df[col].std()

df[num_features].agg(['mean','std']).round(2)
```

## Section 8. Data smoothing (for monthly time-series columns)

To make the monthly billing and payment numbers less jumpy, we use simple exponential smoothing.
Basically, we blend each month's value with the previous one.

$$
x_i^{(smooth)} = 0.1 \times x_{i-1} + 0.9 \times x_i
$$

Where:
- \( x_i \): current month’s amount  
- \( x_{i-1} \): previous month’s amount  
- Coefficients (0.1 and 0.9) control the weight of past vs. present values.

**Why smoothing is used:**

- Reduce short-term random noise in financial activity.
- Highlights long-term behavioral patterns and trends.
- Prevents models from freaking out over one-off weired months.

Smoothing is applied **within each client's time series** across months rather than between clients, because each rows represents diffrent individuals.

---

```python
# Data smoothing across monthly bill and payment amounts
bill_cols = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']
pay_cols  = ['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']

def smooth_series(series, alpha=0.9):
    """
    Applies simple exponential smoothing:
    smoothed[i] = (1 - alpha) * series[i-1] + alpha * series[i]
    """
    smoothed = series.copy()
    for i in range(1, len(series)):
        smoothed.iat[i] = (1 - alpha) * smoothed.iat[i-1] + alpha * smoothed.iat[i]
    return smoothed

# Apply smoothing along each customer's time dimension
for colset in [bill_cols, pay_cols]:
    for col in colset[1:]:  # start from second month because we use previous month
        df[col] = 0.1 * df[colset[colset.index(col) - 1]] + 0.9 * df[col]
```

## Section 11. Mathematical formulas and algorithm justification

**1. Logistic Regression (binary classification)**

We model the probability of default as:
$$
P(y=1 \mid X) = \frac{1}{1 + e^{-(\beta_0 + \sum_{i=1}^{n} \beta_i x_i)}}
$$

- Here \(x_i\) are standardized features (so coefficients are comparable).
- The log-odds interpretation: a one-standard-deviation increase in \(x_i\) multiplies the odds by \(e^{\beta_i}\).

**Why Logistic Regression?**
- Interpretability: coefficients indicate direction and relative strength of association (important for a project report).
- Good baseline for binary classification.

**2. Decision Trees and Random Forests**

- Trees partition the space by choosing splits that maximize impurity reduction (Gini or Entropy).
- Random Forests average many trees to improve generalization and reduce variance.

**Impurity measures:**

Gini index:
$$
Gini = 1 - \sum_{k=1}^{C} p_k^2
$$

Entropy:
$$
Entropy = -\sum_{k=1}^{C} p_k \log_2 p_k
$$

**Why tree-based models?**
- Handle non-linearities and interactions.
- Robust to feature scaling and accommodate both categorical and numeric variables.

**Evaluation metrics:**
- Accuracy (simple baseline)
- Precision, Recall, F1-score (critical for imbalanced classes)
- ROC-AUC (probability ranking quality)
- For imbalanced data, use stratified cross-validation, class weights, or resampling (e.g., SMOTE) where appropriate.
